```python
import numpy as np 
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy import wcs

%matplotlib inline
```

```python
import sys
sys.path.append('../acalib')
from acalib.io import container as cont
import glob
```

```python
from acalib.core.transform import *
from acalib.core.stack import *
c = cont.Container()
path = '../fits/M1-dss2b.fits'
files = glob.glob('../fits/*.fits')

template = fits.open(path)[0].data

for i in files:
    c.load_fits(i)
    

result, snr = stacking(template,c)
```

```python
fig = plt.figure()
fig.add_subplot(111)
plt.imshow(result)
plt.show()
```

```python
import acalib
st = acalib.Stacking()
result2, snr2 = st.run(template,c)
```

```python
fig.add_subplot(121)
plt.imshow(result2)
plt.show()
```
