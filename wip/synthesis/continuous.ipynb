```python
import sympy as sym
from IPython.display import display
%matplotlib notebook
sym.init_printing()
import sympy.stats as sss
x,mu = sym.symbols('x mu')
sigma = sym.symbols('sigma', positive=True)
Ndist=sss.Normal("X",mu,sigma)
f=sss.density(Ndist)(x);f

```

```python
import scipy as sp
from sympy.integrals.transforms import fourier_transform
w=sym.symbols('omega')

cf=fourier_transform(f,x,w).subs({w:(-w/(2*sym.pi)) });cf
```

```python
mgf  = sym.exp(mu*w)*sym.exp(sigma*sigma*w*w/2); mgf
```

```python
i=0
mom=mgf
while i!=20:
    i+=1
    mom = sym.diff(mom,w)
    mm= mom.subs({w:0})
    display(mm)
```
