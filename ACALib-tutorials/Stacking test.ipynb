```python
import numpy as np 
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy import wcs

%matplotlib inline
```

```python
import acalib.io as acaio
import glob
```

```python
c = acaio.Container()
path = '/home/tronco/Downloads/dss_images/messier/m1_070u.fits'
files = glob.glob('/home/tronco/Downloads/dss_images/messier/m*_*.fits')
files = files[:5]
template = fits.open(path)[0].data

for i in files:
    f = acaio.loadFITS_PrimmaryOnly(i)
    c.images.append(f)
```

```python
import acalib
st = acalib.Stacking()
result = st.run(template,c)
```

```python
plt.imshow(result)
```
