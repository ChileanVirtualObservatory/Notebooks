```python
import urllib.request
url="https://raw.githubusercontent.com/carosaa/Laconeu/master/exercise.txt"
urllib.request.urlretrieve(url, "data.txt")
```

```python
import numpy as np
data=np.loadtxt('data.txt')
```

```python
data
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
srate=256
x=np.linspace(0.0,1.0,srate)
plt.plot(x,data)
```

```python
from scipy.fftpack import fft, ifft
y  = data[:,0]
yf = fft(y)
```

```python
T=1/srate
xf=np.linspace(0,1/(2*T),srate/2)
plt.bar(xf,np.abs(yf[0:srate/2]))
```

```python
#mask=(xf>30)
#mask=np.hstack((mask,mask[::-1]))
yf[30:srate-30]=0.0
plt.bar(xf,np.abs(yf[0:srate/2]))
#plt.plot(np.abs(yf))
```

```python
xden=ifft(yf)
```

```python
plt.plot(x,xden,'r')
plt.plot(x,data[:,0],'b')
```

```python
#Interactivity
from ipywidgets import interact

def f(deg=0):
    plt.imshow(ndimage.rotate(brain,deg),cmap='gray')
    
interact(f,deg=(0,180))
```

```python
fig=plt.imshow(brain,cmap='gray')
#cont=plt.contourf(brain)
cont=plt.contour(brain)
plt.colorbar(cont)
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
plt.title("Contours levels of Image")
```

```python
#http://www.scipy-lectures.org/advanced/image_processing/
gfilt=ndimage.gaussian_filter(brain, 2)
mfilt=ndimage.median_filter(brain, 4)

plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(brain)#,cmap='gray')
plt.subplot(133)
plt.imshow(gfilt)#,cmap='gray')
plt.subplot(132)
plt.imshow(mfilt)#,cmap='gray')
 
```
