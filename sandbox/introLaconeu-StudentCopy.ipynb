# 1.3 Lambda Functions (Functions on the fly)

Creation of anonymous functions (i.e. functions that are not bound to a name) at runtime, using a construct called "lambda".  

```python
def f(x): 
    return x**2

print(f(5)) 
```

```python
print((lambda x: x**2)(5))

g= lambda x: x**2
print(g(4))
```

```python
def make_incrementor(n): 
    return lambda x: x + n

i5 = make_incrementor(5)
i7 = make_incrementor(7)
print(i5(10))
print(i7(10))
```

The major advantage of using Numpy for handling arrays, is that **Numpy implements all the heavy lifting in C language** which is much faster. This allows you to tap into the powerful C language performance from the comfort of Python.

```python
import numpy as np
```

## ndarray object
The ndarray object is the main object for representing your array. This object can handle multi-dimensional array of any size that your memory can store. The biggest differences between a Python list and a ndarray object are these:

1. ndarray is a fixed size array while list has a dynamic size. When you reshape a ndarray a new object is created with the new shape and the old object is deleted from memory.
2. ndarray allows mathematical and logical operations on complete multi-dimensional arrays. With a Python list you will have to iterate over the sequence which take more time and code.
3. ndarray has homogeneous data type for the complete array while Python list can contain multiple data types within a single array.

Note: You could have multiple data types in a single object that has multiple dimensions.

# 2.4 Indexing 1D

```python
C = np.array(range(100))
print(C, '\n')

print('the first element is:',C[0],'\n') 

print('the last element is:', C[-1], '\n')

print('the penultimate element is:', C[-2], '\n')

print('range:', C[2:21], '\n') #[2,21)

print('other range:', C[0:10:2])
```

# 2.5 Indexing 2D

```python
D = C.reshape(10,10)
print('D:\n', D,'\n')

print('row 2:',D[2],'\n')

print('column 2:', D[:,2],'\n')

print('(2,3):\n',D[2][3],'\n')

print('(2,3):\n',D[2,3],'\n')

print('matrix range:[1:5,2:4]:\n',D[1:5,2:4],'\n')

print('Even numbers:\n', D[D%2==0],'\n')

print('D > 34:\n', D[D>34])
```

# 2.6 Applying a Function

## You have to vectorize the function

The purpose of np.vectorize is to transform functions which are not numpy-aware (e.g. take floats as input and return floats as output) into functions that can operate on (and return) numpy arrays

1. Note that not every function automatically works with arrays.

2. Functions that have been written to work with arrays of numbers are called vectorized functions.

3. Most of the functions in numpy are already vectorized.

4. You can create a vectorized version of any other function using the higher-order function numpy.vectorize().

```python
def isprime(n):
    #checking the input 
    n= abs(int(n))
    if n<2: #if n=0 or n=1 then return is not a prime
        return False
    if n==2:
        return True
    for x in range(2,n):
        if n % x == 0:
            return False
    return True

```

```python
print(isprime(27), '\n')
```

```python
aux=isprime(D)
```

```python
visprime = np.vectorize(isprime)

tmp = visprime(D)

print(tmp, '\n')

print(D[tmp])
```

# 4.3 Plotting multi-dimensional with matrices
If we supply a matrix to plot() then it will plot the y-values taken from the columns of the matrix (notice the transpose in the example below).

```python
x = np.arange(0, 2*np.pi, 0.01)
y = np.sin(x)
ax = plt.plot(x,np.matrix([y, np.cos(x)]).T)
```
